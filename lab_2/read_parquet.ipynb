```python
from pyspark.sql import SparkSession

# Initialize a SparkSession
# For local execution, use "local[*]"
spark = SparkSession.builder.appName("ReadParquetFile").getOrCreate()
```

```python
# Define the path to your Parquet file(s)
# This can be a single file or a directory containing multiple Parquet files
silver_file_path = "./datamart/silver/loan_daily/silver_loan_daily_2023_02_01.parquet"

try:
    # Read the Parquet file into a PySpark DataFrame
    df = spark.read.parquet(silver_file_path)

    # Show the first few rows of the DataFrame
    print("Silver DataFrame loaded successfully:")
    df.show()

    # Print the schema of the DataFrame to see column names and data types
    print("Silver DataFrame Schema:")
    df.printSchema()

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    # Stop the SparkSession
    spark.stop()
```

```python
# Define the path to your Parquet file(s)
# This can be a single file or a directory containing multiple Parquet files
gold_file_path = "./datamart/gold/label_store/gold_label_store_2023_08_01.parquet"
spark = SparkSession.builder.appName("ReadParquetFile").getOrCreate()

try:
    # Read the Parquet file into a PySpark DataFrame
    df = spark.read.parquet(gold_file_path)

    # Show the first few rows of the DataFrame
    print("Gold DataFrame loaded successfully:")
    df.show(100)

    # Print the schema of the DataFrame to see column names and data types
    print("Gold DataFrame Schema:")
    df.printSchema()

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    # Stop the SparkSession
    spark.stop()
```
